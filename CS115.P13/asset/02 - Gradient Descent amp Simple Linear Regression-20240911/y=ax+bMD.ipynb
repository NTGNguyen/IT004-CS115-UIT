```python
import numpy as np
import matplotlib.pyplot as plt
```

## Model y= ax +b

```python
def f(x):
    return 3*x + 2
def model(a,b,x):
    return a*x + b
```

```python
np.random.seed(142857)
x = np.random.uniform(0,1,50)
y = f(x)
noisses = np.random.normal(0,0.3,50)
t = y + noisses
plt.figure(figsize = (5,3))
plt.plot(x,y,'r--',label = '$f_x$')
plt.plot(x,t,'o',label = '$t$')
plt.xlim(0,1)
plt.ylim(0,5)
plt.legend()
plt.show()
```

```python
np.random.seed(142857)
X = np.array((np.ones_like(x),x)).T
```

```python
def model(X,W):
    return X @ W
def gradient(X,W_0,t):
    return 2 * X.T @ (model(X,W_0) - t)
```

```python
W_0 = np.random.randn(2)

alpha = 0.1
nb_of_running = 50
W_list = []

for i in range(nb_of_running):
    W_0 = W_0 -alpha * gradient(X,W_0,t)
```

```python
plt.figure(figsize = (5,3))
plt.plot(x,y,'k:',label = '$f_x$')
plt.plot(x,t,'o',label = '$t$')
plt.plot([0,1],[model([[1,0]],W_0),model([[1,1]],W_0)],'r--',label = 's')
plt.xlim(0,1)
plt.ylim(0,5)
plt.legend()
plt.show()
```
