```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
sns.set_style('darkgrid')
```

```python
def f(x):
    return x*2
```

```python
np.random.seed(42)
x = np.random.uniform(0,1,10)
x
```

```python
t = f(x)
```

```python
t
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,t,'o',label = '$t$')
plt.xlabel('$x$')
plt.ylabel('$t$')
plt.xlim(0,1)
plt.ylim(0,2)
plt.legend()
plt.show()
```

```python
noise = np.random.normal(0,0.3 , x.shape[0])
t = f(x) + noise
t
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,t,'o',label = '$t$')
plt.xlabel('$x$')
plt.ylabel('$t$')
plt.xlim(0,1)
plt.ylim(0,2)
plt.legend()
plt.show()
```

```python
#Define The loss function()
```

```python
def loss(pred,t):
    return np.mean((t-pred)**2)
def model(x,w_0):
    return w_0*x
```

```python
w_s = np.linspace(0,4,1000)
losses = np.array([loss(model(x,w_o),t) for w_o in w_s])
plt.figure(figsize=(5,3))
plt.plot(w_s, losses, 'r--', label='loss')
plt.xlabel('$w_0$')
plt.ylabel('Loss $L(w)$')
plt.legend()
plt.show()
```

```python
def gradient(w_0,x,t):
    return np.mean(x*2*(model(x,w_0)-t))
```

```python
w_0 = np.random.rand()
alpha = 0.9 #define the learning rate

nb_of_intergration = 4
w_loss = [(w_0,loss(model(x,w_0),t))]

for i in range(nb_of_intergration):
    gd = gradient(w_0,x,t)
    w_0 = w_0 - gd*alpha
    w_loss.append(((w_0,loss(model(x,w_0),t))))
```
