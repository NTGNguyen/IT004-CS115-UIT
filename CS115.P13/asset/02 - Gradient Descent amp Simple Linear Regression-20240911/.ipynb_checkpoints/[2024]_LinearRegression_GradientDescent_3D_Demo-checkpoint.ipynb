```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np
```

```python
# The data to fit
m = 20
w0_true = 2
w1_true = 0.5

x = np.linspace(-1,1,m)
noise = np.random.normal(loc=0.0, scale=0.05, size=m)
y = w0_true + w1_true * x + noise
```

```python
noise
```

```python
def loss_function(w0, w1):
    return np.mean((y - model(x, w0, w1))**2)

def model(x, w0, w1):
    return w0 + w1*x
```

```python
n_ticks = 101
w0_grid = np.linspace(-1,4, n_ticks)
w1_grid = np.linspace(-5,6, n_ticks)

X, Y = np.meshgrid(w0_grid, w1_grid)
J_grid = np.zeros((n_ticks, n_ticks))

for i in range(n_ticks):
    for j in range(n_ticks):
        w0 = X[i, j]
        w1 = Y[i, j]
        J_grid[i, j] = loss_function(w0, w1)
```

```python
ax = plt.figure(figsize=(10,10)).add_subplot(projection='3d')
# X, Y, Z = axes3d.get_test_data(0.05)
# Plot the 3D surface
ax.plot_surface(X, Y, J_grid, edgecolor='royalblue', lw=0.5, rstride=8, cstride=8, alpha=0.3)
# Plot projection of the contours for each dimension
ax.contourf(X, Y, J_grid, 15, zdir='z', offset=-5, cmap='coolwarm')

ax.set(xlim=(-1, 4), ylim=(-5, 6), zlim=(-5, 15), xlabel='X', ylabel='Y', zlabel='Z')
plt.show()
```

```python
# Take N steps with learning rate alpha down the steepest gradient,

N = 6
alpha = 0.5

# Initial point for gradient descent
w0_init = 0
w1_init = -4

ws = [np.array((w0_init, w1_init))]
line_data = np.zeros((N, 3))
J = [loss_function(*ws[0])]
line_data[0, 0] = w0_init
line_data[0, 1] = w1_init
line_data[0, 2] = J[-1]
for j in range(N-1):
    last_w = ws[-1]
    this_w = np.empty((2,))
    this_w[0] = last_w[0] - alpha / m * np.sum(
                                    (model(x, *last_w) - y))
    this_w[1] = last_w[1] - alpha / m * np.sum(
                                    (model(x, *last_w) - y) * x)
    line_data[j+1, 0] = this_w[0]
    line_data[j+1, 1] = this_w[1]

    ws.append(this_w)
    J.append(loss_function(*this_w))
    line_data[j+1, 2] = J[-1]
```

```python
print(line_data)
```

```python
ax = plt.figure(figsize=(10,10)).add_subplot(projection='3d')

# Plot the 3D surface
ax.plot_surface(X, Y, J_grid, edgecolor='royalblue', lw=0.5, rstride=8, cstride=8, alpha=0.3)

# Plot projection of the contours for each dimension
contours = ax.contourf(X, Y, J_grid, 15, zdir='z', offset=-5, cmap='coolwarm')


colors = ['blue', 'green', 'magenta', 'cyan', 'orange', 'yellow', 'pink']
ax.scatter( line_data[:, 0], line_data[:, 1], [-5]*N, c=colors[0:N], s=40, lw=0)
ax.scatter( line_data[:, 0], line_data[:, 1], line_data[:, 2], c=colors[0:N], s=40, lw=0)
ax.plot( line_data[:, 0], line_data[:, 1], [-5]*N, color='red', zorder=10)
ax.plot( line_data[:, 0], line_data[:, 1], line_data[:, 2], color='red', zorder=30)

for j in range(0, N):
    ax.text(line_data[j,0]-0.1, line_data[j,1]+0.1, -5, r"({0})".format(j), color='red', zorder=20)
ax.set(xlim=(-1, 4), ylim=(-5, 6), zlim=(-5, 15), xlabel='w0', ylabel='w1', zlabel='Z')

plt.show()
```
