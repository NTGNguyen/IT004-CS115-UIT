## Module: y=a*x + b

```python
plt.figure(figsize=(5,3))
plt.plot([0,2],[f(0),f(2)],'r--',label = '%f_x%')
plt.xlabel('%x%')
plt.ylabel('%y%')
plt.xlim(0,2)
plt.ylim(0,8)
plt.legend()
plt.show()
```

```python
np.random.seed(23521049)
x = np.random.uniform(0,2,20)
y = f(x)
y
```

```python
plt.figure(figsize=(5,3))
plt.plot([0,2.5],[f(0),f(2.5)],'r--',label = '%f_x%')
plt.plot(x,y,'o',label = "%y%")
plt.xlabel('%x%')
plt.ylabel('%y%')
plt.xlim(0,2.5)
plt.ylim(0,10)
plt.legend()
plt.show()
```

## Create noise


```python
np.random.seed(142857)
noisess = np.random.normal(0,0.3,20)
noisess
```

```python
y=y+noisess
```

```python
plt.figure(figsize=(5,3))
plt.plot([0,2.5],[f(0),f(2.5)],'r--',label = '%f_x%')
plt.plot(x,y,'o',label = "%y%")
plt.xlabel('%x%')
plt.ylabel('%y%')
plt.xlim(0,2.5)
plt.ylim(0,10)
plt.legend()
plt.show()
```

##Start gradient Decent

```python
def model(x,a):
    return x*a
```

```python
def loss_func(y,pred):
    return np.mean((y-pred)**2)
```

```python
def gradient(w_0,x,y):
    return np.mean(2*x*(model(x,w_0) - y))
```

```python
w_s = np.linspace(0, 8, num=1000)
losses = np.array([loss_func(model(x,w_0),y) for w_0 in w_s])
```

```python
plt.figure(figsize=(5,3))
plt.plot(w_s,losses,'r--',label = "$f_l$")
plt.xlim(0,8)
plt.ylim(0,16)
plt.legend()
plt.show()
```

```python
np.random.seed(142857)
w_0 = np.random.rand()
np_of_runnning = 100
alpha = 0.5 #learning rate
w_losses_x = [w_0]
w_losses_y = [loss_func(model(x,w_0),y)]
for i in range(np_of_runnning):
    gra =  gradient(w_0,x,y)
    w_0 = w_0 - alpha * gra
    w_losses_x.append(w_0)
    w_losses_y.append(loss_func(model(x,w_0),y))
w_losses_x = np.array(w_losses_x)
w_losses_y = np.array(w_losses_y)
w_losses_x
```

```python
plt.figure(figsize=(5,3))
plt.plot(w_s,losses,'r--',label = "$f_l$")
plt.plot(w_losses_x,w_losses_y,'o',label='%points%')
plt.xlim(0,8)
plt.ylim(0,30)
plt.legend()
plt.show()
```
